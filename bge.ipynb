## only sorting is done on pincode and ascii

```python
#sorting on basis of pincode
df.sort_values(by=['PIN_CODE'], inplace=True)
```

# Loading the embedding model

```python
from sentence_transformers import SentenceTransformer
```

```python
model = SentenceTransformer('BAAI/bge-large-zh')
```

## creating embeddings for address column

```python
embeddings_1 = model.encode(df['Delivery_Desc'], normalize_embeddings=True)
```

```python
embeddings_1
```

# DBSCAN method

```python
!pip install hdbscan
```

```python
import hdbscan

# Fit HDBSCAN model
clusterer = hdbscan.HDBSCAN(min_cluster_size=2, gen_min_span_tree=True)
cluster_labels = clusterer.fit_predict(embeddings_1)
```

```python
cluster_labels
```

### Results

```python
# Create a new column in the original dataframe with the cluster labels
df['Cluster'] = cluster_labels

# Group the dataframe by the Cluster column and create a list of addresses for each cluster
clustered_addresses = df.groupby('Cluster')['Delivery_Desc'].apply(list)

# Print the clustered addresses
for cluster, addresses in clustered_addresses.items():
    print(f'Cluster {cluster}:')
    for address in addresses:
        print(address)
    print('\n')
```

# 1 to 1 comparison

```python
import torch.nn.functional as F
import torch
```

```python
embeddings_1_tensor=torch.tensor(embeddings_1)
```

```python
def clustering(df):
  cluster_id=[]
  c_id=1
  customer_id=df['CUSTOMER_ID'][0]
  address=df['Delivery_Desc'][0]
  pin_code=df['PIN_CODE'][0]
  cluster_id.append([[customer_id,address,pin_code]])
  for i in range(1,len(df)):
    customer_id=df['CUSTOMER_ID'][i]
    address=df['Delivery_Desc'][i]
    pin_code=df['PIN_CODE'][i]

    if(F.cosine_similarity(embeddings_1_tensor[i],embeddings_1_tensor[i-1],dim=0)>0.85):
      cluster_id[-1].append([[c_id,customer_id,address,pin_code]])
    else:
      c_id+=1
      cluster_id.append([[c_id,customer_id,address,pin_code]])

  for i in range(len(cluster_id)):
    if len(cluster_id[i])<2:
      cluster_id[i]=[]

  return cluster_id
```

```python
clus=clustering(df)
```

# N X N comparisons within pincode

```python
import torch.nn.functional as F

def clustering(df, embeddings_1_tensor):
    unique_pincodes = df['PIN_CODE'].unique()
    cluster_id = []
    c_id = 1

    for pincode in unique_pincodes:
        df_pincode = df[df['PIN_CODE'] == pincode]
        cluster = []
        for i in range(len(df_pincode)):
            customer_id = df_pincode['CUSTOMER_ID'].iloc[i]
            address = df_pincode['Delivery_Desc'].iloc[i]
            pin_code = df_pincode['PIN_CODE'].iloc[i]

            if not cluster:
                cluster.append([c_id, customer_id, address, pin_code])
            else:
                for existing_entry in cluster:
                    existing_embedding = embeddings_1_tensor[df[df['CUSTOMER_ID'] == existing_entry[1]].index[0]]
                    current_embedding = embeddings_1_tensor[df_pincode.index[i]]
                    if F.cosine_similarity(current_embedding, existing_embedding, dim=0) > 0.85:
                        cluster.append([c_id, customer_id, address, pin_code])
                        break
                else:
                    c_id += 1
                    cluster_id.append(cluster)
                    cluster = [[c_id, customer_id, address, pin_code]]
        cluster_id.append(cluster)

    for i in range(len(cluster_id)):
        if len(cluster_id[i]) < 2:
            cluster_id[i] = []

    return cluster_id

# Example usage:
# clusters = clustering(df, embeddings_1_tensor)
```

```python
clus=clustering(df,embeddings_1_tensor)
```
